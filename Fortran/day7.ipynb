```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv("orbit.csv")

# Plot orbit trajectory
plt.figure(figsize=(6,6))
plt.plot(data["x"], data["y"], label="Orbit Path")
plt.plot(0, 0, 'yo', label="Central Star")  # Star at origin
plt.xlabel("x (AU)")
plt.ylabel("y (AU)")
plt.title("Orbit Simulation (Euler's Method)")
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.tight_layout()
plt.show()
```
