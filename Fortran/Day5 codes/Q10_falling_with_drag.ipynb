```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("falling_with_drag.csv")

plt.plot(data['T'], data['v(t)'], label="Velocity", color='blue')
plt.plot(data['T'], data['a(t)'], label="Acceleration", color='orange')
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Falling Object with Air Drag")
plt.grid(True)
plt.legend()
plt.show()
```
