```python
# plot_output.py
import matplotlib.pyplot as plt
import pandas as pd

# Load data
data = pd.read_csv("Qs8_output.csv")

# Plot f(x)
plt.figure(figsize=(10, 5))
plt.plot(data['x'], data['f(x)'], label='f(x) = x²', color='blue')
plt.plot(data['x'], data['integral'], label='Integral of f(x)', color='green')

# Labels
plt.xlabel("x")
plt.ylabel("Value")
plt.title("Trapezoidal Integration of f(x) = x²")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
